# 인공지능
지능형 챗봇, 자율주행 차량, 가상 비서 등 전문영역에 인공지능이 대부분의 의사결정을 하고, 경제활동까지 도맡게 되어, 인간은 최소한의 노동력으로도 극대화된 효유적 활동이 가능해질 것이라 전망

 **머신러닝 = 지적작업을 자동화 하려는 노력을 설명하는 열쇠**

현재 업무는 컴퓨터로 진행하나, 과거 업무는 수기로 진행하였으므로 현재에 비해 작업속도는 현저히 느릴수 밖에 없었다.

자크 드 보캉송(Jacques de Vaucanson, 1709년 2월 24일 - 1782년 11월 21일) : 기계공학의 천재, 플루트를 연주하는 사람크기의 조각상(인형)을 완성(처음으로 생명체의 구조와 원리를 적용한 사례)

<br>
Uncanny Valley Theory - **불쾌함의 골자기 이론** (Masahiro Mori, 1970) : 로봇이 인간을 어설프게 닮을수록 오히려 불쾌감이 증가  
<br>
오토마타를 계기로 산업혁명기의 새로운 소재와 다양한 아이디어로 저마다 인간의 육체와 정신을 모방하는 모험에 합류  
현재의 인공지능 혹은 머신러닝의 초석을 다짐  
<br>
그리고 100년 후 빅토리아시대
세계 최초의 범용 기계식 계산기(차분기관)가 발명됨.

차분기관을 발명한 찰스 바비지의 친구이자 협력자인 영국인 Augusta Ada King, Countess of Lovelace(1815년 12월 10일~1852년 11월 27일)
그는 1843년 이 발명품에 대하여 이렇게 이야기 함.  
"차분 기관은 무엇이든 가리지 않는다.
차분 기관이 어떤 일을 수행하게 하려고 할때
우리가 차분기관에 어떻게 명령해야 하는지를 알고 있다면,
차분기관은 그러한 일을 모두 수행할 수 있다."  
"차분기관의 역할은 우리에게 이미 익숙한일을 가능하게 하는데 있다."  

<br>
Alan Mathison Turing - 앨런 튜링(1912년 6월 23일~1954년 6월 7일) : 인공지능의 서구자중 한 명

**튜링테스트**를 소개한 그의 기념비적인 논문 'Computing machinery and intelligence'에서 러브레이스의 발언을 인용, 법용계산기가 학습을 하거나 독창성을 발휘할 수 있는지에 대하여 깊이 생각해본 후, 그는 이러한 결론을 내리게 된다.
"범용 계산기는 학습과 독창성을 발휘할 수 있다."  
<br>
머신러닝의 근본적인 질문("수행 방법을 아는 모든 일을 넘어서 지정한 작업을 수행하는 방법을 계산기(컴퓨터)가 스스로 학습할 수 있는가?")은 프로그래밍의 새로운 패러다임을 열러줌게 됨.  
<br>
영화 '이미테이션 게임' 중 기계학습에 대한 엘런 튜링의 대사 - "인간의 두뇌는 큰수의 셈도 빨리해내죠. 크리스토퍼(기계)도 셈을 한 후 어떻게 처리할지도 결정하죠. 사람처럼요 전기 두뇌라고 생각하면 돼요. 디지털 컴퓨터죠"  
<br>
고전프로그래밍(기호주의 인공지능)에서 인간은 인간이 설게한 규칙을 입력할 뿐 아니라, 이 규칙에 따라 처리할 데이터를 입력하고 그에 대한 답변을 획득합니다. 즉, 규칙에 의거한 정답을 얻어내기 위한 과정이죠.  
<br>
머신러닝은 사람이 데이터 뿐 아니라 데이터에서 예상되는 해답까지 입렭한 후, 규칙을 뽑아내는 구조로 설계되며, 이러한 구조로 뽑아낸 규칙을 신규 데이터에 적용 인간이 아닌 컴퓨터가 정의한 규칙을 통해 해답이 산출됩니다.  
<br>
머신러닝은 명시적인 프로그래밍이 아닌 훈련의 과정을 진행하게 되며, 주어진 임무와 관련된 다수의 데이터를 접한 후, 이 데이터 속에서 통계적 구조를 찾아내어 임무 완료 과정을 자동화하는데 필요한 규칙을 제시  
<br>
머신러닝은 1990년대에 이르러서야 번성하기 시작했지만, 고속 처리 하드웨어와 대규모 데이터셋을 활용할 수 있게 되면서 급격하게 인공지능 하위 분야 중 가장 인기를 끌게 되었습니다.  
<br>
그럼 머신러닝을 이해하기 위해선 통계학만을 전적으로 공부하면 되는 것일까요?  
대규모 데이터 셋을 통해 그 속에 숨겨진 규칙을 찾아내는 것이라면 데이터 마이닝 또한 데이터 셋을 정제하고, 그 속에 숨겨진 패턴을 찾는 것이니 우린 통계학과 더불어 데이터 마이닝만을 공부하면 되는 것 일까요?  
<br>
머신러닝은 통계학과 밀접한 관련이 있지만, 몇 가지 점에서 통계학과는 다른 면이 있습니다.  
머신러닝은 통계학과 달리 베이즈 분석과 같은 고전적인 통계 분석에 비교적 효율적이지 못하며,
통계학에서 사용되는 데이터 보다 더욱 복잡하면서도 대규모의 데이터셋을 처리해야 하죠.  
<br>
그래서 머신러닝, 더불어 딥러닝을 포함하여 상대적으로 적은 수학 이론만을 표출하며, 오히려 공학적인 측면을 더 지향할때가 많습니다.  

그리고 더 중요한 점은 머신러닝 분야에서는 하나의 아이디어에 대해 <u>'이론'보다는 '경험'을 통해 입증되는 경우가 많다는 것인데(**이론 < 경험**)<u/>,  

실제로 <u>신경망의 정확도 향상 및 과적합을 방지하는데 효과적인 **드롭아웃 기법**<u/>이 탄생한 배경은 제프리 힌튼의 번뜩이는 아이디어에서 시작해 다수의 실험을 거쳐 탄생하게 되었죠.

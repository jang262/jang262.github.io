## 경사하강법  
**경사하강법**은 말 그대로  
<u>목적함수의 값을 최소화 시키기 위해 마치 경사를 내려가듯 최소값을 찾는 기법</u>이에요.  
근데 이 경사하강법은 **미분** 이라는 계념을 집고 넘어가야 합니다.

가로측을 경과시간 세로축을 주행거리라고 하면

이 그래프를  보면 40초에 120m정도를 주행했고, 이를 바탕으로  
이를 바탕으로 자동차의 주행 속도롤 계산해 보면 3m/s가 나옵니다.  
그럼 이 차는 처음 시동을 걸 때부터 끝까지 이 속도로 달린 걸까요?  
아니죠 이건 평균 속도에요.  
자동차는 처음 시동을 걸 때와, 속도를 올릴 때, 중간에 정차했을 때,  
다시 속도를 올릴 때 즉 각각의 시점에서 그 순간순간의 속도는 모두 다른 값이죠.  
앞서 40초간의 속도를 계산했는데 이런 순간의 변화령을 구하기 위해서 점점 간격을 좁혀 보겠습니다.

즉, <u>미분은 어떤 구간에서의 그래프의 기울기를 구하는 것과 같다고 말할 수 있습니다.</u>

10초인 시점에서의 순간의 기울기, 즉 속도를 구할 수 있게 되죠.  
이런식으로 간격을 좁혀서 기울기를 구하는 작업을 미분이라고 말합니다.  

이 순간의 변화량을 구하기 위해 함수를 f(x)라 주고, h를 매우 작은 수라고 하면

이렇게 계산할 수 있겠죠.  
그럼 여기 이 6이라는 값이 기울기고, 이것은 속도가 됩니다.  
사실 h가 한없이 0에 가까워져야 하므로 0.1보다 더 작은 값이어야 하지만,  
이건 하나의 예시니깐 h는 0.1이라 가정하고 계산한 겂니다.  
<br>
그래서 선구자들은 이러한 미분을 통해 목적함수의 최소가 되는 x값을 찾고자  
막 열심히 문제를 푸는데 어.. 먼가 규칙적인 움직임이 보이는거에요.  
그리고 이를 조금더 계산을 해보니깐  
미분에는 몇가지 성질이 있다는 것을 알게 됬죠.  
<img src="https://user-images.githubusercontent.com/54702627/84496162-7050fc80-ace7-11ea-9bb2-6a851eb1a3f2.png" width="450px" height="300px" title="경사하강법7" alt="RubberDuck"></img><br/>
대표적으로 x와 관계 없는 상수를 미분하게 되면 0이 되는것 처럼  
이러한 성질들을 알아두면 언젠가 분명히 여러분의 힘이 될것입니다.  
<br>  

자 그럼 다시 경사하강법으로 돌아오기에 앞서 한가지 실험을 해보죠.  

예를 들어 여기 이러한 2차 함수가 있다고 가정했을 때,  
이 함수의 초소값은 x가 1일때 g(x)는 0 즉 최소값에 위치하게 됩니다.
<img src="https://user-images.githubusercontent.com/54702627/84496505-0422c880-ace8-11ea-8371-6aed79f3d930.png" width="450px" height="300px" title="경사하강법8" alt="RubberDuck"></img><br/>

그럼 이 2차함수를 증감표로 나타내 볼까요?  
앞서 배운 미분의 성질을 떠올려 봅시다.  
<img src="https://user-images.githubusercontent.com/54702627/84497403-ab542f80-ace9-11ea-8d71-d06f55faf97a.png" width="450px" height="300px" title="경사하강법9" alt="RubberDuck"></img><br/>
우선 이 식을 전개하면 x<sup>2</sup> - 2x + 1 이 됩니다.  
그리고 이 식을 미분하면? 이러한 결과(2x - 2)가 나타나겠네요.  
이때 이 식이 미분 후에 나온 함수라 하여, 도함수라 합니다.  
그럼 이 도함수의 부호에 주목하면서 천천히 살펴보죠.  
우선 x의 범위를 양수, 음수, 최소값으로 구분합니다. 여기서 최소값은 1이죠?  
그리고 이 각각의 상황에 맞게 도함수에 대입하여 결과를 살펴보면  
음수일 때는 마이너스, 양수일 때는 플러스  
그리고 최소값일 땐 0이란 부호를 얻을 수 있게 됩니다.  
그럼 이 증감표에 따르면 x가 음수일 때는 그래프가 오른쪽 아래로 내려가고,  
반대로 x가 양수일 땐 오른쪽 위로 올라가는데, 이건 최소값을 찾는 거니 이건 최소값을 찾는거니 이건 왼쪽 아래로 내려간다 할 수 있겠네요.  

자 아주 단순하게 생각해야 됩니다. 

예를 들어 x가 3인 순간부터 시작해서 g(x)의 값을 작게 만들려면  
<img src="https://user-images.githubusercontent.com/54702627/84497622-1f8ed300-acea-11ea-9f6d-8bd1b044b86a.png" width="450px" height="300px" title="경사하강법10" alt="RubberDuck"></img><br/>
x를  왼쪽으로 밀어 옮기면, 즉 x를 작게 만들면 되겠죠?  

반대로 반대쪽에 위치한 음수인 x즉, -1인 순간부터 시작한다면  
<img src="https://user-images.githubusercontent.com/54702627/84497875-84e2c400-acea-11ea-8e1e-ff3535f23bf5.png" width="450px" height="300px" title="경사하강법11" alt="RubberDuck"></img><br/>
g(x)의 값을 작게 만들기 위해서는 x의 값을 오른쪽으로 밀어 옮기면 즉 x를 크게 만들면 되겠죠?  

자 여기서 알 수 있는건 우리가 구한 도함수의 부호에 따라  
x를 밀어 옮기는 방향이 달라지게 된다는 거고,  
쉽게 말해 도함수의 부호 반대 방향으로 밀어 옮기게 되면  
자연스럽게 최소값 쪽으로 움직인다는 의미죠. 
<img src="https://user-images.githubusercontent.com/54702627/84498198-210ccb00-aceb-11ea-84ce-024a1cf759b0.png" width="450px" height="300px" title="경사하강법12" alt="RubberDuck"></img><br/>
이렇게 되면 우리가 구하고자 하는 매개변수들을 일일이 비교하며 구하지 않더라도,  
자동으로 갱신할 수 있다는 의미가 됩니다.  
이러한 이야기를 식으로 정리하면 이와같은 식으로 도출할 수 있습니다.  
<img src="https://user-images.githubusercontent.com/54702627/84498531-ca53c100-aceb-11ea-9d6e-daca8ced7747.png" width="450px" height="300px" title="경사하강법13" alt="RubberDuck"></img><br/>
그리고 이를 경사하강법 이라 부르죠.  
여기 이(:=) 기호는 A를 B에 따라 정의한다는 의미를 가지고 있습니다.  
그리고 또 중요한 개념이 등장하는데 바로 **학습률**입니다.
학습률은 에타 라는 그리스 문자를 사용하며, 양의 정수를 사용합니다.  
이 <u>학습률에 따라 최소 값에 도달하기까지 갱신해야 하는 횟수가 달라지며,  
흔히 수렴되는 속도가 달라진다 라고 표현</u>해요.  

또다시 어려운 수식이 나왔네요.  
이럴때 역시 구체적인 값을 대입하면서 생각해 보는 수 밖에 없죠.  

예를들어 우리 학습율을 1로 정하고 x가 3에서 시작 한다고 했을 때, x는 어떻게 움직일까요?  
<img src="https://user-images.githubusercontent.com/54702627/84499001-a5138280-acec-11ea-9569-548f5b8fe041.png" width="450px" height="300px" title="경사하강법14" alt="RubberDuck"></img><br/>
다시 g(x) 함수를 미분하면 2x-2라는 식이 나오니깐 갱신식은 이렇게 되겠죠?  
그럼 우린 이걸 계산하면 되는 겁니다. 자 계산과 함께 그래프를 잘 봐주세요.  

x값이 최소값에 수렴하지 않고 계속 3과 -1을 반복하는 현상이 나타납니다.  
<img src="https://user-images.githubusercontent.com/54702627/84499499-af824c00-aced-11ea-924a-91694ab4817f.png" width="450px" height="300px" title="경사하강법15" alt="RubberDuck"></img><br/>
최소값을 찾기 위해 경사를 내려가듯이 조심스럽게 내려가야 하는데,  
발걸음이 너무나도 커서 최소값을 찾지 못하는 상황인 거죠.  

그럼 다시 경사를조심스럽게 내려가기 위해 보폭을 줄여보겠습니다.  
<img src="https://user-images.githubusercontent.com/54702627/84499741-1e5fa500-acee-11ea-8c56-e4b78778f9a1.png" width="450px" height="300px" title="경사하강법16" alt="RubberDuck"></img><br/>
학습률을 줄여본다는 의미에요.  
0.1로 설정하고 계산을 진행해보면,  
아까와는 달리 점차 최소값에 수렴해가는 모습을 확인할 수 있습니다.  
하지만 수렴하는 속도가 생각보다 느리죠.  

정리하면 이렇게 학습률의 값을 너무 크게 잡게 되면  
오히려 최소값에서 멀어지는 **'발산'**이 일어나고,  
<img src="https://user-images.githubusercontent.com/54702627/84500046-b9f11580-acee-11ea-964a-cde2411fb3df.png" width="450px" height="300px" title="경사하강법17" alt="RubberDuck"></img><br/>
반대로 너무 작게 잡으면 최소값에 수렴하기 까지 너무 많은 시간이 걸리게 됩니다.  

즉 학습률은 정답이 없으며 앞에 영상에서 이야기한 머신러닝의 실험적 성격이 들어나게 되는 순간이죠.

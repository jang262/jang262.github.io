# 딥러닝

머신러닝 알고리즘이 '하는 일'
기존 입력 데이터와 예상 출력사례를 통해 그 사이의 관계를 기계가 스스로 학습을 통해 찾아낸 시스템이 있고, 이 시스템의 입력값 기준에 맞게 새로운 사례들을 제시하면 머신러닝 과정을 통해 데이터 처리 작업을 실행하게 됩니다.  
따라서 머신러닝을 하기 위해서는 다음 세가지가 필요한 것을 알수 있죠  
<br>

### 머신러닝을 위한 3가지
1. 입력 데이터 :  
예를들어 음성인식 작업인 경우 이데이터점들은 사람의 목소리가 담긴 사운드 파일일 수 있고, 이미지에 태그를 다는 작업인 경우라면 사진이 될 수 있습니다.  
<br>
2. 예상 출력값 :  
음성인식 작업인 경우 사운드 파일을 듣고, 원고 형태로 옮긴 것일 수도 있고, 이미지작업의 경우라면 '강아지'나 '고양이' 등과 같은 태그일 수 있습니다.  
<br>
3. 알고리즘 정확도 측정법 :  
알고리즘의 현재 출력과 예상출력 사이의 거리를 결정하는데 필요하죠.  
이 측정치는 알고리즘 작동 방식을 조정하는 피드백 신호로 사용되며 일반적으로 이러한 조정 단계를 '학습'이라 합니다.  
따라서 머신러닝과 딥러닝의 핵심 과제는 데이터를 의미있는 것으로 변환하는 것이죠.
예상하는 결과에 더 가깝게 접근 할 수 있도록 하는 표현.  
즉, 입력 데이터의 유용한 표현을 학습하는 것 입니다.  
<br>
('표현' 이란 :  
<br>
데이터를 대표하게 하거나 데이터를부호화 하기 위해 데이터를 색다르게 살펴보는 방법.
<br>
예를 들어 컬러이미지를 적색, 녹색, 청색의 RGB형식 또는 색조, 채도, 값의 HSV형식으로 부호화 한다고 생각하면 됩니다.  
<br>
이 두가지 형식은 동일한 데이터를 두가지 방식으로 표현한 것 인데, 특정 표현 방식을 썼을때는 수행하기 어려웠던 작업일지라도 표현방식을 바꾸는 것 만으로도 수행하기 쉬운 작업이 되기도 합니다.  
<br>
예를 들어 모든 적색 픽셀을 선택하는 작업은 HSV 형식보단 RGB형식일때 더 간단하게 처리할 수 있습니다.)  

머신러닝 모델 :  
입력 데이터에 대한 적절한 표현을 찾는 일,  
즉 분류작업과 같이 현재 직면하고 있는 작업에 더 적합한 데이터 변환 방식을 찾는일과 관련.  

X축 Y축 (X,Y)좌표계에서 좌표로 표현되는 몇가지 흰색점과 검은색점이 있다고 생각해 보자.
점의 좌표(X,Y)를 취해 그 점이 검은색인지, 흰색인지를 출력하는 알고리즘을 개발하려 한다고 가정하면,  

* 3가지 가정을 가진다.
1. 입력값은 점들의 좌표이다.
2. 기대하는 결과는 점들의 색이다.
3. 알고리즘이 올바른 작업을 수행하는지 여부를 측정하는 방법은 예를 들어 올바르게 분류된 점들의 백분율일 수 있다.

여기서 우리가 필요로 하는 것은 흰색점과 검은색점을 깔끔하게 구분하는 데이터의 새로운 표현이다.

머신러닝의 맥락에서 학습이란 :
더 나은 표현을 자동으로 찾아내는 과정이라고 설명할 수 있다.

###딥러닝이란 :  
머신러닝의 특정 하위 분야로, 표현들을 점점 더 의미있게 만들어가는 연속 계층들을 학습하게 하는 데 중점을 두고 데이터 표현을 학습하는 새로운 방법.

딥러닝에서 딥이라는 문구는 이러한 접근법에 의해 만들어진.  
즉, '연속된 표현 층' 이라는 개념을 나타낸 약자이다

딥러닝에서 이러한 계층적 표현은 **'신경망'**이라고 부르는 모델을 통해 학습되며, 각기 서로 포개지기도 하고 교차하기도 하며 다양한 방식(**교차/반복/결합** 등)으로 구조화 되어 있다.  
이러한 **'신경망'**이라는 용어는 신경 생물학에서 차용한 것 이고, 딥러닝의 중심 개념 중 일부는 뇌를 이해한 데서 얻은 영감을 통해 부분적으로 개발되었다.
그래서 신경망을 설명하려면 우선 생물학에서 다루는 뉴런을 이야기 해야 한다.

* 뉴런
1. 뉴런은 네트워크를 형성한다.
2. 여러 누런에 전달되는 신호의 합이 일정크기(**임곗값**)을 넘지 않으면 뉴런은 전혀 반응하지 않는다.
3. 여러 누런에 전달되는 신호의 합이 일정 크기 **임계값**을 넘으면 뉴런이 반응하며, 다른 뉴런에 일정 강도의 신호를 전달한다.
4. 2와 3에서 여러 뉴런에 전달되는 신호의 합은 신호마다 가중치가 다르다.

이러한 뉴런의 작용을 수학적으로 추상화해서 일정 단위의 인위적인 네트워크로 표현한 것이 **신경망**이다.
즉, 뇌를 형성하는 뉴런의 집합체를 수학 모델로 나타내는 것이 신경망의 출발점이며 앞서 이야기 했듯 이러한 신경망의 계층을 다양한 방식으로 연속되게 구조화 한 방식이 딥러닝이라 할 수 있다.
쉽게 이야기 하면 데이터 표현들을 여러 단계를 거쳐 배우는 방식이라 생각하면 된다.

###딥러닝은 어떻게 학습을 이루는가?

입력x -> 계층(데이터 변환) -> 다수의 계츨(데이터 변환) -> 예측y  

입력 데이터는 다음의 해당 계층에 의해 데이터가 특정화 된 다음 이 계층의 '가중치'라는 형태로 저장
이 가중치는 이후 살펴보겠지만, 본질적으로 여러 숫자로 된 덩어리이다.
그리고 이후 다양한 계층들을 반복적으로 통과하며 학습을 진행한다.
이러한 맥락에서 생각할때 '학습'이란 신경망에서 모든 계층의 가중치에 대한 값 집합을 찾아 새로운 입력값이 목표한 값에 이르기까지 가중치를 조절하는 반복적업이라 생각할 수 있다.
이때 심층 신경망에서는 수천만개의 가중치가 존재하며, 모든 가중치들을 대상으로 올바른 값을 찾는 일은 그리 쉽지 않을 것이라는 의구심이 들 수 있습니다. 
특히 이러한 가중치는 하나의 값을 수정하면 다른 모든 가중치에 영향을 주기에 더욱 그렇다.

조금 더 넓은 관점으로 살펴보면
신경망의 출력을 제어하려면 이 출력이 예상한 것과 얼마나 차이가 나는지를 측정할 수 있어야 하는데 그렇게 위해선 다양한 방법이 있는데 우선 가장 기본적인 부분부터 살펴보자.
**목적 함수**라고 부르는 심경망의 **손실함수(loss function)** : 
손실함수는 실제 결과값과 신경망이 예측한 결과값을 가지고 거리 점수를 계산하여 이 특정 사례에서 신경망이 얼마나 잘 수행되었는지 계산해 낸다.

딥러닝의 기본 기법은 이 점수를 피드백 신호로 사용해 현재 사례의 **손실 점수**를 낮추는 방향으로 무작위 가중치값들을 조금씩 조절한다.
이러한 조절 방식은 주로 **역전파 알고리즘**을 통해 구현하는데, 이후 강의에서 **역전파**에 대하여 자세하게 설명하도록 하고, 우선은 가볍게 신경망의 **손실점수**를 낮춰주는 하나의 방식이라 생각하면 되지 않을까 합니다.

최종적으로 손실이 최소화된 신경망은 우리가 원하는 값에 가장 근접한 출력을 낼 수 있으며 이는 곧 가장 잘 훈련된 신경망이라 할 수 있습니다.
